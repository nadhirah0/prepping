```python
import pandas as pd
import numpy as np
import os
```

```python
pwd = os.getcwd()
```

```python
dataset1 = pd.read_csv('Transactions.csv')
dataset1
```

```python
dataset2 = pd.read_csv('Swift_Codes.csv')
dataset2
```

```python
dataset_modified1 = dataset1.copy()
dataset_modified1
```

```python
dataset_modified2 = dataset2.copy()
dataset_modified2
```

```python
(dataset_modified1.dtypes)
```

```python
dataset_modified2= dataset_modified2.assign(Country_Code=['GB','GB','GB','GB','GB','GB','GB'])
dataset_modified2
```

```python
dataset_modified1[['Sort Code_']] = dataset_modified1[['Sort Code']].astype(pd.StringDtype())
dataset_modified1
```

```python
(dataset_modified1.dtypes)
```

```python
dataset_modified1[['Sort Code_1','Sort Code_2','Sort Code_3']] = dataset_modified1['Sort Code_'].str.split("-", expand = True)
dataset_modified1
```

```python
dataset_modified1['Sort Code_'] = dataset_modified1['Sort Code_1'].map(str) + dataset_modified1['Sort Code_2'].map(str) + dataset_modified1['Sort Code_3'].map(str)
dataset_modified1
```

```python
dataset_comb = pd.merge(dataset_modified1, 
                      dataset_modified2, 
                      on ='Bank', 
                      how ='outer')
dataset_comb
```

```python
dataset_comb.drop(dataset_comb.columns[[5, 6, 7]], axis=1, inplace=True)
dataset_comb
```

```python
(dataset_comb.dtypes)
```

```python
dataset_comb[['Account Number']] = dataset_comb[['Account Number']].astype('str')
dataset_comb
```

```python
(dataset_comb.dtypes)
```

```python
dataset_comb['IBAN'] = dataset_comb['Country_Code'] + dataset_comb['Check Digits'] + dataset_comb['SWIFT code'] + dataset_comb['Sort Code_'] + dataset_comb['Account Number']
dataset_comb
```

```python
output=dataset_comb[['Transaction ID','IBAN']]
output
```

```python
# determining the name of the file
file_name = 'output.xlsx'
  
# saving the excel
output.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```
