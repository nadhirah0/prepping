```python
import pandas as pd
import numpy as np
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "/table1.xlsx")
dataset
```

```python
dataset_modified = dataset.copy()
dataset_modified
```

```python
relative_rank = dataset_modified.groupby('Month')['Value'].rank(ascending= False)
relative_rank = dataset_modified.assign(relative_rank = relative_rank )
relative_rank
```

```python
table2 = relative_rank.sort_values(['Month', 'relative_rank'],ascending = [True, True])
table2
```

```python
perRank = pd.pivot_table(table2, values='Value', index='relative_rank', aggfunc=np.mean)
#perRank = table2.groupby('relative_rank')['Value'].mean()
perRank
```

```python
perBank = pd.pivot_table(table2, values='relative_rank', index='Bank', aggfunc=np.mean)
#perBank = table2.groupby('Bank')['relative_rank'].mean()
perBank
```

```python
dataset_comb = pd.merge(table2, 
                      perRank, 
                      on ='relative_rank', 
                      how ='inner')
dataset_comb
```

```python
dataset_comb2 = pd.merge(dataset_comb, 
                      perBank, 
                      on ='Bank', 
                      how ='outer')
dataset_comb2
```

```python
dataset_comb2 = dataset_comb2.sort_values(['Month', 'relative_rank_x'],ascending = [True, True])
dataset_comb2
```

```python
dataset_comb3 = dataset_comb2.set_axis(['Month', 'Bank', 'Value', 'relative_rank', 'PERBANK', 'PERRANK'], axis=1, copy=False)
dataset_comb3
```

```python
# determining the name of the file
file_name = 'dataave.xlsx'
  
# saving the excel
dataset_comb3.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```
