```python
dataset = pd.read_csv('DSB Customer Survery.csv')
dataset
```

```python
dataset_modified = dataset.copy()
dataset_modified
```

```python
id_vars = list(dataset_modified.columns)[ : 1]
value_vars = list(dataset_modified.columns)[1 : ]
```

```python
dataset_melted = dataset_modified.melt(id_vars=id_vars, value_vars = value_vars)
dataset_melted
```

```python
# split column into multiple columns by first delimiter 
dataset_melted[['Platform','Cat']]=dataset_melted['variable'].str.split(" - ", n=1, expand = True)
dataset_melted
```

```python
dataset_melted = dataset_melted[['Customer ID','Platform','Cat','value','variable']]
dataset_melted
```

```python
dataset_melted = dataset_melted.sort_values(['Customer ID','Platform'])
dataset_melted
```

```python
dataset_melted = dataset_melted.set_index("Cat")
dataset_melted = dataset_melted.drop("Overall Rating")
dataset_melted
```

```python
perPlatform = pd.pivot_table(dataset_melted, values='value', index=['Customer ID','Platform'], aggfunc=np.mean).reset_index()
#perPlatform = dataset_melted.groupby(['Customer ID','Platform'])['value'].mean()
#perPlatform = perPlatform.reset_index()
perPlatform
```

```python
def combo_subtraction(perPlatform, level=0):
    unique_groups = perPlatform.columns.levels[level]
    combos = itertools.combinations(unique_groups, 2)
    
    pieces = {}
    for g1, g2 in combos:
        name = "{}-{}".format(g1, g2)
        pieces[name] = perPlatform.xs(g1, level=level, axis=1) - perPlatform.xs(g2, level=level, axis=1)
    return pd.concat(pieces)

out = (perPlatform.pivot(index="Customer ID", columns="Platform") # convert data to wide format
       .pipe(combo_subtraction, level=1)       # apply our combination subtraction
       .dropna()                               # clean up the result
       .swaplevel()
       .sort_index()
       .reset_index())

out
```

```python
def set_CatCust(row):
    if row["value"] == "0":
        return "mobile_g"
    else:
        return "none"

out['CatCust'] = out.apply(set_CatCust, axis=1)
out
```
