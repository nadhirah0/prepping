```python
import pandas as pd
import numpy as np
import os
```

```python
pwd = os.getcwd()
```

```python
# parse_dates read transaction date column as datetime values, correctly taking the first part of the date input as the day.
# (ddmmyyyy format) 
dataset = pd.read_csv('PD 2023 Wk 1 Input.csv', parse_dates=['Transaction Date'], dayfirst=True)
dataset
```

```python
dataset_modified = dataset.copy()
dataset_modified
```

```python
# split column into multiple columns by first delimiter 
dataset_modified[['Bank','Transaction']]=dataset_modified['Transaction Code'].str.split("-", n=1, expand = True)
dataset_modified
```

```python
# new column utk month baesd on column date 
dataset_modified['Month'] = pd.to_datetime(dataset_modified['Transaction Date']).dt.strftime('%m')
dataset_modified
```

```python
table1 = pd.pivot_table(dataset_modified, values='Value', index=['Month','Bank'], aggfunc=sum)
table1
```

```python
table2 = table1.sort_values(['Month', 'Value'],ascending = [True, False])
table2
```

```python
# determining the name of the file
file_name = 'table1.xlsx'
  
# saving the excel
table1.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```
