```python
dataset2 = pd.read_csv('Account Information.csv')
dataset2
```

```python
dataset_modified2 = dataset2.copy()
dataset_modified2
```

```python
(dataset_modified2.dtypes)
```

```python
dataset_modified2[['Account1','Account2']]=dataset_modified2['Account Holder ID'].str.split(", ", n=1, expand = True)
dataset_modified2
```

```python
id_vars = list(dataset_modified2.columns)[ : 5]
value_vars = list(dataset_modified2.columns)[5 : ]
```

```python
dataset_melted = dataset_modified2.melt(id_vars=id_vars, value_vars = value_vars)
dataset_melted
```

```python
dataset_melted = dataset_melted.dropna(subset=['value']).drop(['variable','Account Holder ID'], axis=1)
dataset_melted
```

```python
dataset_melted = dataset_melted.rename(columns = {'value':'Account Number'})
dataset_melted
```
