```python
import pandas as pd
import numpy as np
import os
```

```python
pwd = os.getcwd()
```

```python
# parse_dates read transaction date column as datetime values, correctly taking the first part of the date input as the day.
# (ddmmyyyy format) 
dataset1 = pd.read_csv('Week_3.csv', parse_dates=['Transaction Date'], dayfirst=True)
dataset1
```

```python
dataset2 = pd.read_csv('Targets.csv')
dataset2
```

```python
dataset_modified1 = dataset1.copy()
dataset_modified1
```

```python
dataset_modified2 = dataset2.copy()
dataset_modified2
```

```python
(dataset_modified1.dtypes)
```

```python
(dataset_modified2.dtypes)
```

```python
# split column into multiple columns by first delimiter 
dataset_modified1[['Bank','Transaction']]=dataset_modified1['Transaction Code'].str.split("-", n=1, expand = True)
dataset_modified1
```

```python
dataset_modified1['Quarter'] = pd.to_datetime(dataset_modified1['Transaction Date']).dt.quarter
dataset_modified1
```

```python
table1 = pd.pivot_table(dataset_modified1, values='Value', index=['Bank'], aggfunc=np.sum)
table1
```

```python
dataset_dsb=dataset_modified1[dataset_modified1['Bank'] != 'DS']
dataset_dsb
```

```python
dataset_dsb=dataset_dsb[dataset_dsb['Bank'] != 'DTB']
dataset_dsb
```

```python
dataset_dsb[['Online or In-Person']] = dataset_dsb[['Online or In-Person']].astype('str')
dataset_dsb.head()
```

```python
(dataset_dsb.dtypes)
```

```python
dataset_dsb['Online or In-Person'] = dataset_dsb['Online or In-Person'].replace({'1':'Online', '2':'InPerson'})
dataset_dsb
```

```python
table2 = pd.pivot_table(dataset_dsb, values='Value', index=['Online or In-Person','Quarter'], aggfunc=np.sum)
table2
```
